```python

# =====================================================================
# 1. LOAD DATA (Maps to: dairy.txt)
# =====================================================================
print("Loading 'dairy.xlsx'...")


# Maps to: LOGGED DEPENDENT VARIABLE (Y/N) -> y
y = df["ly"].values

# Maps to: NUMBER OF REGRESSOR VARIABLES (Xs) -> 4
# We manually add the intercept to make it 5 columns total.
X = np.column_stack([
    np.ones(len(df)),    # Intercept (Beta 0)
    df["lland"].values,  # Beta 1
    df["llabor"].values, # Beta 2
    df["lfeed"].values,  # Beta 3
    df["lcattle"].values # Beta 4
])

# =====================================================================
# 2. FIT MODEL (Maps to: MU=n, ETA=n, TIME=1, PRODUCTION=1)
# =====================================================================
print("Fitting CrossSectionalSFA (Half-Normal Production Model)...\n")

# model_type="production" handles the s=1 logic
# CrossSectionalSFA inherently enforces mu=0 and eta=0
model = CrossSectionalSFA(model_type="production")
model.fit(X, y)

# =====================================================================
# 3. PRINT RESULTS (Maps to: TE_SFA1.txt)
# =====================================================================
print("="*55)
print(f"{'Parameter':<15} | {'PanelSFA':<15} | {'Frontier 4.1':<15}")
print("="*55)

# True values from your previous Frontier 4.1 output
frontier_vals = {
    "Beta 0": 7.7252632,
    "Beta 1": 0.03603705,
    "Beta 2": 0.10265138,
    "Beta 3": 0.15563085,
    "Beta 4": 0.75468361,
    "Sigma-sq": 0.048396738,
    "Gamma": 0.90051465,
    "Log-Likelihood": 110.46602,
}

# Use index for Intercept, for Land, etc.
print(f"{'Beta 0':<15} | {model.coef_:<15.8f} | {frontier_vals['Beta 0']:<15.8f}")
print(f"{'Beta 1':<15} | {model.coef_:<15.8f} | {frontier_vals['Beta 1']:<15.8f}")
print(f"{'Beta 2':<15} | {model.coef_:<15.8f} | {frontier_vals['Beta 2']:<15.8f}")
print(f"{'Beta 3':<15} | {model.coef_:<15.8f} | {frontier_vals['Beta 3']:<15.8f}")
print(f"{'Beta 4':<15} | {model.coef_:<15.8f} | {frontier_vals['Beta 4']:<15.8f}")
print(f"{'Sigma-sq':<15} | {model.sigma_sq_:<15.8f} | {frontier_vals['Sigma-sq']:<15.8f}")
print(f"{'Gamma':<15} | {model.gamma_:<15.8f} | {frontier_vals['Gamma']:<15.8f}")
print(f"{'Log-Likelihood':<15} | {model.log_likelihood_:<15.8f} | {frontier_vals['Log-Likelihood']:<15.8f}")
print("="*55)

# =====================================================================
# 4. TECHNICAL EFFICIENCY ESTIMATES
# =====================================================================
df['TE_Score'] = model.score_efficiency(X, y)
mean_te = df['TE_Score'].mean()
frontier_mean_te = 0.85325139

print(f"\nMean Efficiency:  {mean_te:.8f} (Frontier 4.1: {frontier_mean_te:.8f})")
```

```python

# Helper to prevent formatting errors if parameters are arrays
def to_float(val):
    return float(val) if isinstance(val, (np.ndarray, list)) else float(val)

# =====================================================================
# 1. LOAD DATA


y = df["ly"].values
X = np.column_stack([
    np.ones(len(df)),    # Beta 0
    df["lland"].values,  # Beta 1
    df["llabor"].values, # Beta 2
    df["lfeed"].values,  # Beta 3
    df["lcattle"].values # Beta 4
])

# =====================================================================
# 2. FIT MODEL
# =====================================================================
print("Fitting CrossSectionalSFA...")
model = CrossSectionalSFA(model_type="production")
model.fit(X, y)

# =====================================================================
# 3. ROBUST COMPARISON TABLE
# =====================================================================
frontier_vals = {
    "Beta 0": 7.7252632,
    "Beta 1": 0.03603705,
    "Beta 2": 0.10265138,
    "Beta 3": 0.15563085,
    "Beta 4": 0.75468361,
    "Sigma-sq": 0.048396738,
    "Gamma": 0.90051465,
    "Log-Likelihood": 110.46602,
}

print("\n" + "="*60)
print(f"{'Parameter':<15} | {'PanelSFA':<20} | {'Frontier 4.1'}")
print("-"*60)

# Loop through coefficients using indices to avoid TypeError
labels = ["Beta 0", "Beta 1", "Beta 2", "Beta 3", "Beta 4"]
for i, label in enumerate(labels):
    val = to_float(model.coef_[i])
    print(f"{label:<15} | {val:<20.8f} | {frontier_vals[label]:.8f}")

# Format variance and likelihood
print(f"{'Sigma-sq':<15} | {to_float(model.sigma_sq_):<20.8f} | {frontier_vals['Sigma-sq']:.8f}")
print(f"{'Gamma':<15} | {to_float(model.gamma_):<20.8f} | {frontier_vals['Gamma']:.8f}")
print(f"{'Log-Likelihood':<15} | {to_float(model.log_likelihood_):<20.8f} | {frontier_vals['Log-Likelihood']:.8f}")
print("="*60)

# =====================================================================
# 4. EFFICIENCY CHECK
# =====================================================================
te_scores = model.score_efficiency(X, y)
print(f"\nMean Efficiency:  {te_scores.mean():.8f}")
print(f"Frontier 4.1 Mean: 0.85325139")
```
